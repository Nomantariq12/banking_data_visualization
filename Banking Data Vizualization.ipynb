# Visualizing Data of Banks

```python
# importing Liabraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Defining Size of Diagrams
plt.rcParams["figure.figsize"] = (20, 10)
```

```python
# Styling for Graphs
plt.style.use("fivethirtyeight")
```

### Investment in PIB's as a Proportion of their own invsetment portfolio

```python
# Different Bank Data
dic = {
    "bop": [14, 0, 58, 38, 71],
    "hbl": [45, 29, 35, 53, 61],
    "akbl": [49, 38, 40, 63, 51],
    "ubl": [45, 48, 56, 20, 21],
    "mcb": [34, 16, 39, 35, 61]
}
dic
```

```python
# Converting Dictionary Data into Pandas DataFrame
# .T is for transpose. So that columns are shifted to indices positions
# The reason for doing this is to plot different bar plots for same bank multiple time. 
data = pd.DataFrame(dic).T

# X-axis data for bar plot, which is just counting of total number of banks we have.
x = np.arange(len(data[0]))
x
```

```python
data
```

```python
# setting the width of the bar plot
width = 0.15
```

```python
#Bar Plot for BOP (Bank of Punjab)
plt.bar(x - (width * 2), data[0], width=width)
#This is the representation of percetagne above each bar plot.
for i in range(len(x)):
    plt.text(i - (width * 2), data.iloc[i, 0] + 1, str(data.iloc[i, 0]) + "%", ha="center", va="baseline",
            color="darkblue")

#Bar Plot for HBL (Habib Bank Limited)
plt.bar(x - width, data[1], width=width)
for i in range(len(x)):
    plt.text(i - width, data.iloc[i, 1] + 1, str(data.iloc[i, 1]) + "%", ha="center", va="baseline", 
            color="red")

#Bar Plot of AKBL (Askari Bank Limited)
plt.bar(x, data[2], width=width)
for i in range(len(x)):
    plt.text(i, data.iloc[i, 2] + 1, str(data.iloc[i, 2]) + "%", ha="center", va="baseline")

#Bar Plot of UBL (United Bank Limited)
plt.bar(x + width, data[3], width=width)
for i in range(len(x)):
    plt.text(i + width, data.iloc[i,3]+1, str(data.iloc[i,3]) + "%", ha="center", va="baseline", color="green")

#Bar Plot of MCB (Muslim Commercial Bank)
plt.bar(x + (width * 2), data[4], width=width)
for i in range(len(x)):
    plt.text(i + (width * 2), data.iloc[i,4]+1, str(data.iloc[i,4])+"%", ha="center", va="baseline",
            color="grey")

# Title, and x-axis labels
plt.title("Investment in PIB's as a Proportion of their own investment portfolio", loc="left")
plt.xlabel("Bank")
# Replacing x Values(that are just numbers) to Bank Names
plt.xticks(x, data.index.str.upper())
# plt.savefig("Investment_In_PIB's.png")
plt.show()
```

# Advances to Deposits Ratio Visualization 

```python
#Bank Data
banks = ["sbl", "jsbl", "bahl", "mebl", "bok", "bafl", "hbl", "hmb", "mcb", "akbl", "abl", "bop", "ubl", "fabl",
        "scbpl", "snbl", "nbp"]
bank = [i.upper() for i in banks]

diff_bet_2021_and_2012 = [29, 23, 12, 12, 10, 7, 2, 0, 0, -3, -9, -13, -14, -15, -19, -26, -27]
len(bank) // 2
```

```python
width1 = 0.5
```

```python
# Bar Plot for Differnt Banks
plt.bar(bank, diff_bet_2021_and_2012, color="#3792cb", width=width1)
for i in range(len(banks) // 2):
    plt.text(i, diff_bet_2021_and_2012[i]-1, str(diff_bet_2021_and_2012[i])+"%",
            va="center",
            ha="center",
            color="white")
for y in range(len(banks) // 2 + 1, len(banks)):
    plt.text(y, diff_bet_2021_and_2012[y]+1, str(diff_bet_2021_and_2012[y])+"%",
            va="center",
            ha="center",
            color="white")
plt.text(7, -1, "0%", color="black", va="center", ha="center")
plt.text(8, -1, "0%", color="black", va="center", ha="center")

plt.title("Change in Bank Strategy - Advances to Deposits Ratio", loc="left")
plt.ylabel("Difference between 2021 and 2012")
plt.xlabel("Bank")
plt.show()
```
